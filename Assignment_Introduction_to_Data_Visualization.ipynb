# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
df.head()
# import matplotlib.pyplot as plt
# import numpy as np

x = np.arange(len(df['month_number']))  # positions for X-axis (0,1,2,...)
width = 0.35  # width of bars

plt.bar(x - width/2, df['facecream'], width=width, label='facecream')
plt.bar(x + width/2, df['facewash'], width=width, label='facewash')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('facecream vs facewash Sales')
plt.xticks(ticks=x, labels=df['month_number'])  # custom X labels
plt.legend()
plt.show()

```

```
# bar-chart for total unit sold in each month for toothpaste and bathing soap
x = np.arange(len(df['month_number']))
width = 0.35

plt.bar(x-width/2,df['toothpaste'],width=width,label='toothpaste')
plt.bar(x+width/2,df['toothpaste'],width=width,label='bathingsoap')
plt.xlabel('Month')
plt.ylabel('Total unit sold')
plt.legend()
plt.show()
```

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here
plt.plot(df['month_number'],df['bathingsoap'],label='bathingsoap')
plt.plot(df['month_number'],df['facecream'],label='facecream')
plt.plot(df['month_number'],df['facewash'],label='facewash')
plt.plot(df['month_number'],df['toothpaste'],label='toothpaste')
plt.plot(df['month_number'],df['shampoo'],label='shampoo')
plt.plot(df['month_number'],df['moisturizer'],label='moisturizer')

plt.title('Sales trend of product over months')
plt.xlabel('Month')
plt.ylabel('Units sold')
plt.legend()
plt.grid(True)
plt.show()


```
