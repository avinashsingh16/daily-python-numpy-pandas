```python
import ast
def convert_to_list(x):
  try:
    return ast.literal_eval(x) if isinstance(x,str) else x
  except Exception as e:
    print("Failed to parse",x)
    return []
```

```python
test_1 = "[1, 2, 3]"
print(convert_to_list(test_1))  # Output: [1, 2, 3]
```

```python
test_2 = "[{'name': 'Tom'}, {'name': 'Jerry'}]"
print(convert_to_list(test_2))
# Output: [{'name': 'Tom'}, {'name': 'Jerry'}]
```

```python
test_3 = "[1, 2, "   # This is an incomplete list (error!)
print(convert_to_list(test_3))
# Output: Failed to parse: [1, 2,
#         []
```

```python
test_4 = [4, 5, 6]
print(convert_to_list(test_4))  # Output: [4, 5, 6]
```

```python
test_5 = "Hello World"
print(convert_to_list(test_5))
# Output: Failed to parse: Hello World
#         []
```

```python
def extract_genres(genre_str):
  try:
    genre_list = ast.literal_eval(genre_str)
    return [g['name'] for g in genre_list]
  except (ValueError,SyntaxError,TypeError):
    return []
```

```python
#Find the average popularity of movies for each genre.
#1 split the genre strings in to lists
imdb_df['genres_list'] = imdb_df['genres'].dropna().apply(lambda x:x.split(', '))
# explode the list to create one row per genre
imdb_exploded = imdb_df.explode('genres_list')
#group by individual genre and calculate mean popularity
genre_popularity = imdb_exploded.groupby('genres_list')['popularity'].mean().sort_values(ascending=False)
print(genre_popularity)
```

```python
# Which year had the highest total revenue from movies?
int(imdb_df.groupby('Year')['revenue'].sum().idxmax())
```

```python
# What is the average runtime for each production country?
imdb_df['production_countries'][0]

import ast

def extract_countries_names(countries):
  try:
    country_list = ast.literal_eval(countries)
    return [c['name'] for c in country_list]
  except:
    return []
#Apply to create a new column
imdb_df['country_list'] = imdb_df['production_countries'].dropna().apply(extract_countries_names)
#explode the list of countries so each country has its own row
exploded_df = imdb_df.explode('country_list')
#group by country and compute avg runtime
avg_runtime_by_country = exploded_df.groupby('country_list')['runtime'].mean()
print(avg_runtime_by_country)
```

```python
# Which language and year combination had the highest average revenue?
imdb_df.groupby(['original_language','Year'])['revenue'].mean().idxmax()
```

```python
	# Which genres had the highest average popularity over the years?

imdb_exploded.groupby(['genres_list','Year'])['popularity'].mean().idxmax()
```

```python
# Which production company has the highest average budget?
import ast
imdb_df['production_companies'][0]
def extract_production_company(company):
  try:
    productn_comp = ast.literal_eval(company)
    return [c['name'] for c in productn_comp]
  except:
    return []

imdb_df['production_company_list'] = imdb_df['production_companies'].dropna().apply(extract_production_company)
imdb_exploded_production_comp = imdb_df.explode('production_company_list')
grouped_prod_com = imdb_exploded_production_comp.groupby('production_company_list')['budget'].mean().idxmax()
print(grouped_prod_com)
```

```python
	# What is the trend of average popularity for each language across the years?


avg_popularity_each_lang = imdb_df.groupby(['Year','original_language'])['popularity'].mean().unstack()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
avg_runtime_per_year.plot(kind='line',marker='o')
plt.title('Average movie runtime over the years')
plt.xlabel('Year')
plt.ylabel('Average Runtime(minutes)')
plt.grid(True)
plt.show()
```
